### Statistika

Studi Kasus Pengujian Hipotesis Mengenai Pemberian Diskon Hipotesis:

**Hipotesis Nol (H0):** Tidak ada perbedaan pelanggan yang menggunakan diskon atau tidak mengggunakan diskon dalam tingkat kepuasan membeli.

**Hipotesis Alternatif (H1):** Pelanggan yang menggunakan diskon memiliki tingkat kepuasan dalam membeli.

```python
data.describe()
```

#### Pengujian hipotesis

```python
from scipy import stats
```

```python
puas = data[data['Satisfaction_Level']=='1']
biasa_saja = data[data['Satisfaction_Level']=='2']
tidak_puas = data[data['Satisfaction_Level']=='3']
```

```python
# Hipotesis Testing
puas_data = df[df['Satisfaction_Level'] == '1']['Discount_Applied']
biasa_saja_data = df[df['Satisfaction_Level'] == '2']['Discount_Applied']
tidak_puas_data = df[df['Satisfaction_Level'] == '3']['Discount_Applied']
```

#### Hitung kalkulasi

```python
# Hitung jumlah konversi
n_success_puas = puas['Discount_Applied'].sum()
n_success_biasa_saja = biasa_saja['Discount_Applied'].sum()
n_success_tidak_puas = tidak_puas['Discount_Applied'].sum()

# Menghitung semua sample pada control dan treatment
n_obs_puas = puas['Discount_Applied'].count()
n_obs_biasa_saja = biasa_saja['Discount_Applied'].count()
n_obs_tidak_puas = tidak_puas['Discount_Applied'].count()
```

```python
print(n_obs_puas, n_success_puas, n_obs_biasa_saja, n_success_biasa_saja, n_obs_tidak_puas, n_success_tidak_puas)
```

#### Hitung Hipotesis

```python
import numpy as np
import statsmodels.stats.proportion as sp

# Pairwise comparisons
pairs = [
    (n_success_puas, n_obs_puas, n_success_biasa_saja, n_obs_biasa_saja),
    (n_success_puas, n_obs_puas, n_success_tidak_puas, n_obs_tidak_puas),
    (n_success_biasa_saja, n_obs_biasa_saja, n_success_tidak_puas, n_obs_tidak_puas)
]

for pair in pairs:
    count = np.array([pair[0], pair[2]])
    nobs = np.array([pair[1], pair[3]])
    stats, pvalue = sp.proportions_ztest(count, nobs)
    print(f"Z-statistic: {stats}, p-value: {pvalue}")
```

```python
print('P-Value : {0:0.4f}'.format(pvalue))
if pvalue >= 0.05:
    print('Failed to reject H0')
else :
    print('Reject H0')
```

## Machine Learning Modelling

### Train Test Split

```python
from sklearn.model_selection import train_test_split

X = df[['Gender', 'Age', 'Average_Rating',
        'Discount_Applied','Items_Purchased']]

y = df['Satisfaction_Level']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.16, random_state=38)
```

```python
len(X_train)
```

```python
len(X_test)
```

### Normalisasi Data

```python
from sklearn.preprocessing import MinMaxScaler
scaller = MinMaxScaler()
X_train = scaller.fit_transform(X_train)
X_test = scaller.transform(X_test)
```

### Machine Learning Modelling

#### Decision Tree Classifier (GridSearchCV) Logistic Regression




```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
dtree = LogisticRegression()
param_grid = {
    'penalty':['l2'],
    'dual':[False],  # Wrap 'False' in a list
    'tol':[0.0001],
    'C':[1,1.5,2,2.5,3],
    'fit_intercept':[True],
    'intercept_scaling':[1],
    'class_weight':[None],
    'random_state':[None],
    'max_iter':[300],
    'multi_class':['auto'],  # Change 'deprecated' to a valid option like 'auto'
    'verbose':[0],
    'warm_start':[False],
    'n_jobs':[None],
    'l1_ratio':[None],
    'solver':['liblinear']

}

# Perform a grid search with cross-validation to find the best hyperparameters
grid_search = GridSearchCV(dtree, param_grid, cv=3)
grid_search.fit(X_train, y_train)

# Print the best hyperparameters
print(grid_search.best_params_)
```
